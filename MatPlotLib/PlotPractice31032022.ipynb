# BAR

```python
w = .3
X = ['python','C','C++','JS','Java']
p1 = [75,24,72,98,92]
p2 = [42,64,36,52,66]
p = np.arange(len(X))
px = [j+w for j in p]


plt.title("Bar Graph",size=20)
plt.xlabel('Languages',size=15)
plt.ylabel('Popularity',size=15)

color=['g','r','b','#993908','#ae45ec']
color1=['#120145','#452896','#859674','#120569','#023695']

plt.bar(p,p1,width=w,color=color)
plt.bar(px,p2,width=w,color=color1)


plt.xticks(p,X,rotation=90)
plt.legend([2008,2018])

plt.show()
```

```python
# HOrizental Bar
w = .3
X = ['python','C','C++','JS','Java']
p1 = [75,24,72,98,92]
p2 = [42,64,36,52,66]
p = np.arange(len(X))
px = [j+w for j in p]


plt.title("Bar Graph",size=20)
plt.ylabel('Languages',size=15)
plt.xlabel('Popularity',size=15)

color=['g','r','b','#993908','#ae45ec']
color1=['#120145','#452896','#859674','#120569','#023695']

plt.barh(p,p1,w,color = color)
plt.barh(px,p2,w,color = color1)


plt.yticks(p,X,rotation=45)
plt.legend([2008,2018])
plt.show()
```
