```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
# %inline matplotlib
```

```python
yarray = np.random.random(100)*100
```

```python
yarray
```

```python
xarray = np.linspace(0,20,100)
xarray
```

```python
zarray = np.random.random(100)*120
zarray
```

```python
plt.plot(xarray,yarray,c='skyblue')
plt.plot(xarray,zarray,c='salmon')
plt.legend(['X-Y Plot','X-Z Plot'],)
plt.title("Testing in MatplotLib")
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
```

```python
plt.scatter(xarray,yarray,c = 'skyblue',alpha=.4)
plt.scatter(xarray,zarray,c= '#c673d1',alpha=.1)
```

```python
x = np.linspace(0,20,100)
y = np.cos(x)
plt.plot(x,y,c='#c673d1',linestyle=':',linewidth=2)
plt.plot(x,np.sin(x),c='red',linestyle='-.',linewidth=1.5)
```

```python
plt.fill_between(x,x**.15,x**0.5,color='green')
```
