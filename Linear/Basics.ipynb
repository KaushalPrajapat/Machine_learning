```python
df = pd.read_csv('test.csv')

df
```

```python

sb.regplot(df.Roll,df.English,color='y')
X = df.Roll
Y = df.English

meanX = np.mean(df.Roll)
meanY = np.mean(df.English)


n=len(df.Roll)

numer=0
denom=0
for i in range(n):
    numer+=(X[i]-meanX)*(Y[i]-meanY)
    denom+=(X[i]-meanX)**2
    
b1 = numer/denom
b0=meanY-(b1*meanX)
print("c = ",b0,"m = ",b1)

# Cofficient of Determination
```
